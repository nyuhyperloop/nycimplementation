# NYC Implementation Team Alternate Locations Research

The initial proposed route for NYU Hyperloop was Lake Grove and Grand Central Terminal. In Fall 2020, the NYC Implementation Analyst Subteam lead by Akshit Modi primarily looked into New Jersey and Riverhead as alternative locations. Research was conducted in September to December 2020 by Alexander Kelso, Alessandra Vennema, Sophie Xu, Alexander Cushen and Akshit Modi.

## Drivers for location selection

**Social and environmental footprint:**  
Routes and stop locations which stand to benefit their local communities and serve their interests are prioritized. Impacts of pollution or gentrification must also be at the forefront of the selection process.  
  
**Centrality:** We want to consider stop locations which are in areas of high population density, especially populations which are likely to be underserved (higher amount of those commuting to NYC than exisiting transportation can handle).  
  
**Accessibility:** Since a hyperloop has an incredibly high throughput, potential stop locations must have the resources and or infrastructure to support large flows of passengers (ie. major roads, public transportation).  
  
**Ease of implementation:** Regulations and costs provide a major barrier to implementation. Locations and routes which can use existing brownfield infrastructure are ideal, as this minimises cost and impact on local communities. Larger cities tend to have more potential ridership but provide more diffculty in terms of legal regulations.

## Grand Central Terminal (Manhattan)

Positives:
* Popular station (2nd busiest MTA station)
* Centrally located in Manhattan
* Big station allows for more foot traffic
* Near many large offices, especially big banks & tech companies
* Connections to many local lines and busses, as well as the Metro-North train

Considerations:
* High implmentation costs, expensive real estate
* Extremely busy station
* Regulatory hurdles would likely be very high (legal and construction barriers)
* LIRR is being built into the station in the next few years

## Smith Haven Mall, Lake Grove (Long Island) 

Positives:
* Centrally located on Long Island - access to major modes of transport within LI while alternative modes of transportation to the city take longer, and/or are more complicated
* Near highly populated cities, including Brookhaven, Islip and Smithtown
* Located in a relatively wealthy/socioeconomically developed area, people would be more willing to pay higher fare for Hyperloop
* Plenty of real estate for development

Why Smith Haven Mall Specifically?  
It is the main attraction for Lake Grove. There is plenty of real estate that could be bought and used to put a Hyperloop station. 6,330 parking spots are available and could be converted, considering that it is more than half the population of Lake Grove.

Considerations:
* Still far away from many popular locations further out on Long Island
* May bring unwelcome tourism/crowds to a relatively small town

## Fulton St (Manhattan)

Positives:
* Main station for offices on/near Wall St (Lots of Wall St bankers have homes in LI)
* Located in the heart of lower Manhattan
* Access to many main local subway lines, which go uptown as well as into Brooklyn
* Relatively large station, could accommodate more foot traffic
* Connected to Cortlandt St station, WTC station, and Chamber St station, NJ transit system, and Westfield mall inside Oculus
  
Considerations:
* Located far from everything else (ie. Midtown, Uptown, Queens, etc)
* Trajectory of path from LI station to Fulton St would cross directly through many residential and commercial areas in Brooklyn
* Like Grand Central and 34 St Herald Sq, implementation costs and regulatory hurdles would likely be very high
* That area is also typically extremely crowded, not only with people but also with buildings (even for NYC)

## Riverhead (Long Island)

Overview: Riverhead is a good location for additional expansion, but not inital establishment of a Hyperloop system. Commutership from Riverhead to Manhatton is hard to estimate but generally fairly low. In comparison to other locations, Riverhead attracts less riders but also leads to less issues with accommodating commuters, adhering to regulations and maintenance costs. It is a good location for an additional stop after establishing a system between Lake Grove and Grand Central Terminal.

Population: 33,539 (2018)  
Population density: 500 people per square mile - for reference, Long Island is 2,360 and Lake Grove is 3,726  
  
Attractions: LI Aquarium, Tanger Outlets, Jamesport Farm Brewery, Riverhead Raceway, Suffolk Theater  
Characteristics: suburban, low-crime, small, peaceful, decent schools [Official Town Website](https://www.townofriverheadny.gov/)  

Median annual household income: 72,134 USD - for reference, US average is 55,322 USD and Lake Grove is 105,709 USD  
  
Rough estimation of 10k commuters a day to Riverhead area in total. The Ronkonkoma Branch is the main option for timely access to Manhattan with an annual ridership of about 9,800,000 or about 26,700 daily. Ronkonkoma station is also right next to Lake Grove.

## New Jersey

Overview: A stop in New Jersey would need to be sufficiently far from Manhattan to justify implementing a Hyperloop system and  to assure that the Hyperloop can accelerate properly. Many cities in NJ are too close to Manhattan and have too many existing options for commuters. New Brunswick is a site for potential additional expansion.

### Existing services in Newark

Existing services include the NJ Transit lines, PATH and Amtrakâ€™s Northeast Corridor.  
  
Stations are currently undergoing 18 million dollars in renovations. It is the third busiest station on the Northeast Corridor line, which stretches from Boston to Washington, D.C. The station is 10 minutes from Newark airport, which served more than 40.5 million total passengers in 2016. It would be unlikely to be able to build a straight line through Newark for the Hyperloop.  
According to New York Metropolitan Transportation Council (NYMTC) estimates, Manhattan-bound subways carried 389,000 passengers from New Jersey during the peak hour of the morning commute (8:00 AM) on an average fall work day into the Manhattan central business district. The distance of 14 miles is not enough for the Hyperloop to accelerate.  
  
Inner NJ Locations served include Bergen, Essex, Hudson, Middlesex, Morris, Passaic, Somerset, and Union  
Outer NJ Locations served include Hunterdon, Mercer, Monmouth, Ocean, Sussex, and Warren  
  
Conclusion: A NJ station positions us for long term expansion, but too many alternative transportation methods are available and the distance is too short for Newark to be a viable option.
